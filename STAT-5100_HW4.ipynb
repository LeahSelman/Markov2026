```r
install.packages("markovchain")
library(markovchain)
```

```r
# Problem 2c

states = c("0", "1", "2", "3", "4", "5")

p = matrix(c(0,0,0,0,0,1, 0.1,0.9,0,0,0,0, 0,0.1,0.9,0,0,0, 0,0,0.1,0.9,0,0, 0,0,0,0.1,0.9,0, 0,0,0,0,0.1,0.9), nrow=6, ncol=6, byrow=TRUE,
        dimnames = list(c("0", "1", "2", "3", "4", "5"),
        c("0", "1", "2", "3", "4", "5")))


markov_chain = new("markovchain", states = states, transitionMatrix = p)
markov_chain

final_state = steadyStates(object = markov_chain)
final_state

cat("The probability of only one machine working on any given week is: ", final_state[2],"\n")
```
