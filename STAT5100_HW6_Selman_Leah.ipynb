```r
install.packages("markovchain")
library(markovchain)
```

```r
#Problem 1b

states = c("0", "1", "2", "3", "4")

p = matrix(c(0,1,0,0,0, 1/3,0,2/3,0,0, 0,1/2,0,1/2,0, 0,0,2/3,0,1/3, 0,0,0,1,0), nrow=5, ncol=5, byrow=TRUE,
        dimnames = list(c("0", "1", "2", "3", "4"),
        c("0", "1", "2", "3", "4")))


markov_chain = new("markovchain", states = states, transitionMatrix = p)
markov_chain

stat_dist = steadyStates(object = markov_chain)
stat_dist
```

```r
#Problem 4b)
set.seed(95)

a = 0.49

count = vector(,length=1000)
for (i in 1:1000){
    count[i] = 0
    X = 1
    while (X > 0){
        Z = rbinom(1,X, a)
        X = 2*Z
        count[i] = count[i]+1
        }
}

cat("The fraction of avalanches that go extinct before 200 generations is ", sum(count<200)/1000,"\n")

```
