```r
# Problem 2f

states = c("G", "S", "D")

p = matrix(c(9/10,0,2/5,1/10,7/8,0,0,1/8,3/5), nrow=3, ncol=3,
        dimnames = list(c("G", "S", "D"),
        c("G", "S", "D")))
#p

markov_chain = new("markovchain", states = states, transitionMatrix = p)
#markov_chain


N = 10000
simulate = rmarkovchain(n = N, object = markov_chain, t0 = "G")

table(simulate)

cat("The fraction of time spent in state G is: ", sum(simulate=="G")/N,"\n")


#steadyStates(markov_chain)
```

```r
# Problem 4c

states = c("1", "2", "3", "4", "5", "6")

p = matrix(c(1/2,0,1/3,0,0,0, 1/2,1/2,0,0,0,0, 0,1/2,1/3,0,0,0, 0,0,1/3,1/2,0,0, 0,0,0,1/2,0,1, 0,0,0,0,1,0), nrow=6, ncol=6,
        dimnames = list(c("1", "2", "3", "4", "5", "6"),
        c("1", "2", "3", "4", "5", "6")))
#p

markov_chain = new("markovchain", states = states, transitionMatrix = p)
#markov_chain

end_state = matrix(NA, nrow = 10000, ncol = 1)
for (i in 1:10000){
    N = 5
    simulate = rmarkovchain(n = N, object = markov_chain, t0 = "1")
    end_state[i] = simulate[5]
    }

sum(end_state == "4")
cat("The fraction of simulations in which the end state was 4 is: ", sum(end_state == "4")/10000,"\n")
```
