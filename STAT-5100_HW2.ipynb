```r
# Problem 1d

gamma = 4
x0 = 10
N1 = 100
N2 = 1000
N3 = 10000

# generate random values U~U(0,1)
U1 = runif(N1, 0, 1)
U2 = runif(N2, 0, 1)
U3 = runif(N3, 0, 1)

# find x using F^-1
x1 = x0 / ((1-U1)^(1/(gamma-1)))
x2 = x0 / ((1-U2)^(1/(gamma-1)))
x3 = x0 / ((1-U3)^(1/(gamma-1)))

#plot histogram with N=100
hist(x1, freq=FALSE, xlim = c(10,60), main = "N=100", breaks = 100)
curve((gamma-1)*x0^(gamma-1)*x^(-gamma), col = 2, lty = 2, lwd = 2, add = TRUE)

#plot histogram with N=1000
hist(x2, freq=FALSE, xlim = c(10,60), main = "N=1000", breaks = 100)
curve((gamma-1)*x0^(gamma-1)*x^(-gamma), col = 2, lty = 2, lwd = 2, add = TRUE)

#plot histogram with N=10000
hist(x3, freq=FALSE, xlim = c(10,60), main = "N=10000", breaks = 100)
curve((gamma-1)*x0^(gamma-1)*x^(-gamma), col = 2, lty = 2, lwd = 2, add = TRUE)
```

```r
# Problem 2b

start_time = Sys.time()
options(warn=-1)

N = 10^6
U = runif(N, 0, 1)

#F = function(u){F = -exp(-u) * (u+1) + 1; return(F)}

F = -exp(-U) * (U+1) + 1


x_inv = matrix(NA, nrow = N, ncol = 1)
for (i in 1:N){
    find_zero = function(x){zero = x-F[i]; return(zero)}
    x_inv[i] = uniroot(find_zero, c(0,1))
}

end_time = Sys.time()
cat("The runtime of this code is:", end_time-start_time, " minutes")
```

```r
# Problem 2c

c = 1.472

fx = function(x){f = x*exp(-x); return(f)} #pdf
gx = function(x){g = 0.5*exp(-x/2); return(g)} #proposed function

start_time = Sys.time()

n = 0
accepted = c()
while (n < 10^6){
    x = runif(1, 0, 1)
    U = runif(1,0,1)

    if (U < (fx(x)/(c*gx(x)))){
        accepted = c(accepted, x)
        n = n+1
    }
}


end_time = Sys.time()
cat("The runtime of this code is:", end_time-start_time, " minutes")
```
